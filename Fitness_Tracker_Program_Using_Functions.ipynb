```python
# Fitness Tracker Program Using Functions

# Global variable (accessible throughout the program)
user_name = "David"
weekly_goal = 70000  # Weekly step goal


# Function to calculate weekly average steps
def calculate_weekly_average(daily_steps):
    # Local variables
    total_steps = sum(daily_steps)
    days = len(daily_steps)

    average = total_steps / days
    return average


# Function to check if daily goal is achieved
def check_daily_goal(steps, daily_goal):
    # Local variable
    goal_met = steps >= daily_goal

    if goal_met:
        return "Goal achieved!"
    else:
        return "Goal not achieved."


# Function to calculate remaining steps for weekly goal
def check_weekly_progress(total_steps):
    # Local variable
    remaining_steps = weekly_goal - total_steps

    if remaining_steps > 0:
        return f"{remaining_steps} more steps needed to reach weekly goal."
    else:
        return "Weekly goal completed!"


# Function to display fitness summary
def display_summary(name, steps_data):
    average = calculate_weekly_average(steps_data)
    total = sum(steps_data)

    print("----- Fitness Summary -----")
    print("User:", name)
    print("Total weekly steps:", total)
    print("Average daily steps:", round(average, 2))
    print(check_weekly_progress(total))


# Fitness data for 7 days
daily_steps = [8500, 10000, 7500, 12000, 9000, 11000, 9500]

# Calling functions
display_summary(user_name, daily_steps)

print("\nDaily Goal Check:")
for day, steps in enumerate(daily_steps, start=1):
    print("Day", day, ":", steps, "steps -", check_daily_goal(steps, 10000))
```
