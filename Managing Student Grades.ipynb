```python
# Question 1: Print each student's name using a for loop

students = ["Alice", "Ben", "Chloe", "David"]

print("Student Names:")
for student in students:
    print(student)
```

```python
# Question 2: Enter grades until the teacher types "done"

grades = []

while True:
    grade = input("Enter a student's grade (or type 'done' to finish): ")

    if grade.lower() == "done":
        break

    grades.append(float(grade))

print("\nGrades Entered:")
print(grades)

if len(grades) > 0:
    average = sum(grades) / len(grades)
    print("Average Grade:", average)
else:
    print("No grades were entered.")
```

```python
# Question 3: Demonstrate break and continue

grades = []

while True:
    entry = input("Enter a grade (or type 'done' to stop): ")

    # Stop entering grades
    if entry.lower() == "done":
        print("Grade entry stopped.")
        break

    grade = float(entry)

    # Skip invalid grades
    if grade < 0:
        print("Invalid grade! Negative grades are not allowed.")
        continue

    grades.append(grade)
    print("Grade added successfully.")

print("\nValid Grades:")
print(grades)

if len(grades) > 0:
    average = sum(grades) / len(grades)
    print("Average Grade:", average)
else:
    print("No valid grades entered.")
```

```python
# Question 4: Print student names organized by class using nested loops

classes = [["Alice", "Ben"], ["Chloe", "David"]]

for class_number, classroom in enumerate(classes, start=1):
    print(f"\nClass {class_number}")

    for student in classroom:
        print(student)
```
